```python
# Display the last 5 rows of the DataFrame.
import pandas as pd
file_path = 'train_u6lujuX_CVtuZ9i.csv'
df = pd.read_csv(file_path)
print(df.tail())
```

```python
import pandas as pd

#1.Identifying Columns with missing values
missing_columns = df.columns[df.isnull().any()]

#2.Number of missing values in each columns
missing_counts = df[missing_columns].isnull().sum()

#Result
print("Columns with missing values",missing_columns)
print("Number of missing values in each column",missing_counts)
```

```python
# List all the columns which contain non-numerical entries
import pandas as pd

non_numerical_columns = df.select_dtypes(exclude=['number']).columns
print("Columns with non numeric entries : ",non_numerical_columns)
```

```python
''' Find all the Loan IDs of people who are not graduates.
    Find all the Loan IDs of people who have an Applicant Income of less than 5000.'''
import pandas as pd

non_graduates = df[df['Education']=='Not Graduate']
loan_ids_nongrad = non_graduates['Loan_ID']
print("Load ids of people who are non graduates : \n",loan_ids_nongrad)

low_income_applicants = df[df['ApplicantIncome'] < 5000]
loan_ids_lowinc = low_income_applicants['Loan_ID']
print("\nLoan ids of aplicants with income less than 5000 : \n",loan_ids_lowinc)
```

```python
# Find all the Loan IDs where the person is unmarried and has dependents '3+'.
import pandas as pd

marital_status_dep = df[(df['Married']=='No') & (df['Dependents'] == '3+')]
loanids_unmarried_dep = marital_status_dep['Loan_ID']
print("Loan ids of the people who are unmarried and has dependents '3+':\n",loanids_unmarried_dep)
```

```python
# Print all the rows which have Dependents '3+'.
import pandas as pd

dependents_3plus = df[df['Dependents']=='3+']
print("Rows with Dependents 3+ are : \n ",dependents_3plus)
```

```python
# Drop the rows having null/missing values.
import pandas as pd

df_cleaned = df.dropna()
print("Dataframe after dropping rows with null values : \n",df_cleaned)
```
